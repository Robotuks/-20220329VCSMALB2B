```python
import pandas as pd
import numpy as np

# from sklearn import 
from sklearn.preprocessing import OneHotEncoder, RobustScaler
```

```python
df = pd.read_csv('penguins_lter.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df_modified = df.drop( columns=['Region', 'Stage', 'Individual ID', 'Date Egg', 'Comments', 'studyName', 'Sample Number', 'Delta 15 N (o/oo)', 'Delta 13 C (o/oo)'])
df_modified.head()
```

```python
df_modified.tail()
```

# Missing values

```python
df_modified[ df_modified.isna().any(axis=1) ]
```

```python
print (df_modified.shape)
df_modified = df_modified.loc[ df_modified['Culmen Length (mm)'].notnull() ]
print (df_modified.shape)
```

```python
print (df_modified.shape)
df_modified = df_modified.loc[ df_modified['Sex'].notnull() ].reset_index(drop=True)
print (df_modified.shape)
```

```python
df_modified[ df_modified.isna().any(axis=1) ]
```

# Categorical data encoding

```python
df_modified.head()
```

```python
df_modified["Clutch Completion"] = np.where(df_modified["Clutch Completion"] == "Yes", 1, 0)
df_modified.head()
```

```python
df_modified["Sex"] = np.where(df_modified["Sex"] == "MALE", 1, 0)
df_modified.head()
```

```python
encoder = OneHotEncoder()
encoded = encoder.fit_transform( df_modified[ ['Species', 'Island' ] ] )
encoded_col = encoder.get_feature_names_out(['Species', 'Island' ])
print(encoded.shape)
print (encoded_col.shape)
# encoded_df = pd.DataFrame.from_records(encoded.toarray() )
encoded_df = pd.DataFrame(data=encoded.toarray(), columns=encoded_col.tolist())
encoded_df
```

```python
df_modified = df_modified.join( encoded_df)
df_modified
```

```python
df_modified = df_modified.drop(columns= ['Species', 'Island'])
df_modified
```

# Data normalization

```python
data_transformer = RobustScaler()
df_normalized = df_modified.copy()
normalized_cols = data_transformer.fit_transform( df_normalized[ ['Culmen Length (mm)', 'Culmen Depth (mm)', 'Flipper Length (mm)', 'Body Mass (g)'] ] )
df_normalized.loc[:, ['Culmen Length (mm)', 'Culmen Depth (mm)', 'Flipper Length (mm)', 'Body Mass (g)'] ] = normalized_cols
```

```python
df_normalized
```

# Training

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import xgboost as xgb

X_train, X_test, y_train, y_test = train_test_split(df_normalized[ df_normalized.columns.difference(['Sex']) ], df_normalized['Sex'], test_size=0.2, random_state=15)

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
print ('knn', knn.score(X_test, y_test) )

xgb_model = xgb.XGBRFClassifier()
xgb_model.fit(X_train, y_train)
print ('xbg', xgb_model.score(X_test, y_test) )
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_modified[ df_modified.columns.difference(['Sex']) ], df_modified['Sex'], test_size=0.2, random_state=15)

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
print ('knn', knn.score(X_test, y_test) )

xgb_model = xgb.XGBRFClassifier()
xgb_model.fit(X_train, y_train)
print ('xbg', xgb_model.score(X_test, y_test) )
```

# Cross-validation

```python
from sklearn.model_selection import cross_val_score, KFold

kf = KFold(n_splits=5, random_state=15, shuffle=True)
knn = KNeighborsClassifier(n_neighbors=7)

accs = cross_val_score(estimator=knn, X=df_normalized[ df_normalized.columns.difference(['Sex']) ], y=df_normalized['Sex'], cv=kf)
print (accs)
acc = accs.mean()
print (acc)
```

```python
xgb_model = xgb.XGBRFClassifier(use_label_encoder=False)
accs = cross_val_score(estimator=xgb_model, X=df_normalized[ df_normalized.columns.difference(['Sex']) ], y=df_normalized['Sex'], cv=kf)
print (accs)
acc = accs.mean()
print (acc)
```

# Hyperparameter-tuning 

```python
from sklearn.model_selection import GridSearchCV
parameters = {'weights' : ('uniform', 'distance'), 'n_neighbors':[3, 5, 7, 9]}

knn = KNeighborsClassifier()

clf = GridSearchCV(knn, parameters)
clf.fit(X=df_normalized[ df_normalized.columns.difference(['Sex']) ], y=df_normalized['Sex'])

# GridSearchCV(estimator=knn,
#              param_grid={'C': [1, 10], 'kernel': ('linear', 'rbf')})
```
