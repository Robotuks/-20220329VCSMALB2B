# Clustering

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate sample data
centers = [[1, 1], [-1, -1], [1, -1]]
X, Y = make_blobs(n_samples=1000, centers=centers, cluster_std=0.4)
```

```python
plt.scatter(X[:, 0], X[:, 1])
```

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
clusters = kmeans.labels_
print (clusters.shape)
```

```python
clusters
```

```python
plt.scatter(X[:, 0], X[:, 1], c=clusters, label=clusters)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_predictions(Y, clusters)
```

```python


point = [-1,0]
point_class = kmeans.predict( [ point ] )
print (point_class)
colors = ['m', 'y', 'b']
plt.scatter(X[:, 0], X[:, 1], c=clusters)
plt.scatter(point[0], point[1], c=colors[point_class[0]])
plt.scatter(point[0], point[1], c='r', s=1)
```

```python
from itertools import cycle
color_list = list("bgrcmykbgrcmykbgrcmykbgrcmyk")
colors = cycle(color_list)

labels_unique = np.unique(clusters)
n_clusters_ = len(labels_unique)

for k, col in zip(range(n_clusters_), colors):
    my_members = clusters == k
    # cluster_center = cluster_centers[k]
    plt.scatter(X[my_members, 0], X[my_members, 1], c=col)

point = [0,0]
point_class = kmeans.predict( [ point ] )

plt.scatter(point[0], point[1], c=color_list[point_class[0]])
plt.scatter(point[0], point[1], c='y', s=2)
```

# PCA

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.decomposition import PCA


iris = datasets.load_iris()
X = iris.data
Y = iris.target
```

```python
X.shape
```

```python
X_reduced = PCA(n_components=3).fit_transform(X)
X_reduced.shape
```

```python
fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)
ax.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    X_reduced[:, 2],
    c=Y,
    cmap=plt.cm.Set1,
    edgecolor="k",
    s=40
)
```

```python
X_reduced = PCA(n_components=2).fit_transform(X)
X_reduced.shape
```

```python
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=Y, cmap=plt.cm.Set1, edgecolor="k")
```
